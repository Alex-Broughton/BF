```python
def getData(repo, col, fluxkey="base_CircularApertureFlux_70_0_instFlux", shapekey="base_SdssShape_xx", craig_cuts=True, 
            spatial_cuts=True, yousuke_cuts=True):
    
    # Pulls out data from run and returns BF table
    
    subbutler = Butler(repo,collections=col)
    subregistry = subbutler.registry
    
    datasetRefs = list(subregistry.queryDatasets(
        datasetType="gridSpotSrc",
        collections=col))

    tab = []
    for i, aref in enumerate(datasetRefs):

        # postISRCCD
        md = subbutler.get("raw",dataId=aref.dataId).getMetadata()
        exptime = md["EXPTIME"]

        src = subbutler.get("gridSpotSrc",dataId=aref.dataId)
        mdsrc = src.getMetadata()
        x0 = mdsrc['GRID_X0']
        y0 = mdsrc['GRID_Y0']
        
        src = np.where(src['spotgrid_index'] >= 0)
        
        if len(src) == 0:
            continue
        
        # Craig shape cuts
        if craig_cuts:
            
            minSizeX = 4.8
            maxSizeX = 5.1
            minSizeY = 4.8
            maxSizeY = 5.1
            minFluxRatio = 0.80
            maxFlux = np.nanmax(src['base_SdssShape_instFlux'])
            select = src['base_SdssShape_instFlux'] > maxFlux * minFluxRatio
            spotCatalog  = src.subset(select)
            select = ((src['base_SdssShape_xx'] >= minSizeX) & (src['base_SdssShape_xx'] <= maxSizeX) & 
                    (src['base_SdssShape_yy'] >= minSizeY) & (src['base_SdssShape_yy'] <= maxSizeY))

            src  = src.subset(select)
        
        if len(src) == 0:
            continue
        
        # Spatial Cuts
        maxradius = 10.*np.mean([mdsrc["GRID_XSTEP"], mdsrc["GRID_YSTEP"]])
        distances = np.sqrt((src["spotgrid_x"] - x0)**2 + (src["spotgrid_y"] - y0)**2)
        mask1 = (distances <= maxradius)
        
        # Yousuke Cuts
        Iperr = src["base_PsfFlux_instFluxErr"]
        Ip = src["base_PsfFlux_instFlux"]    
        Ierr = src["{}Err".format(fluxkey)]
        I = src["{}".format(fluxkey)]            

        mask2 = (src["base_PixelFlags_flag"] == False) & \
                (src["base_PsfFlux_flag_edge"] == False) & \
                (src["base_CircularApertureFlux_70_0_flag_apertureTruncated"] == False) & \
                (src["base_ClassificationExtendedness_flag"] == False) & \
                (src["base_PixelFlags_flag"] == False) & \
                (src["base_PixelFlags_flag_offimage"] == False) & \
                (src["base_PixelFlags_flag_edge"] == False) & \
                (src["base_PixelFlags_flag_interpolated"] == False) & \
                (src["base_PixelFlags_flag_saturated"] == False) & \
                (src["base_PixelFlags_flag_cr"] == False) & \
                (src["base_PixelFlags_flag_bad"] == False) & \
                (src["base_PixelFlags_flag_suspect"] == False) & \
                (src["base_PixelFlags_flag_interpolatedCenter"] == False) & \
                (src["base_PixelFlags_flag_saturatedCenter"] == False) & \
                (src["base_PixelFlags_flag_crCenter"] == False) & \
                (src["base_PixelFlags_flag_suspectCenter"] == False) & \
                (Iperr/Ip<0.01 ) & \
                (Ierr/I<0.01)

        
        if spatial_cuts and yousuke_cuts:
            src = src[mask1 & mask2]
        elif spatial_cuts and not yousuke_cuts:
            src = src[mask1]
        elif not spatial_cuts and yousuke_cuts:
            src = src[mask2]
        else:
            continue
            
        if len(src) > 0:
            tab.append(
                {
                    "index": i,
                    "obsId": md["OBSID"],
                    "exptime": exptime,
                    "numspots" : len(src),
                    "spot_indices": src['spotgrid_index'],
                    "spotgrid_x": src['spotgrid_x'],
                    "spotgrid_y": src['spotgrid_y'],
                    "shape": src[shapekey],
                    "flux": src[fluxkey]
                }
            )

    t = Table(tab)
    t.sort("obsId")
    outputTable = vstack(t, join_type='exact').group_by('exptime')
    
    return outputTable

```

```python
# Write data to a pickle

byamp_results = []
byamp_corrected_results = []


for i, table in enumerate([tab_corr, tab_uncorr]):
    for group in table.groups:

        flux = []
        shape = []
        for exp in group:
            flux.extend(exp['flux'])
            shape.extend(exp['shape'])

        if i==0:
            byamp_corrected_results.append([np.sum(group['numspots']),np.nanmean(flux), np.nanstd(flux), np.nanmean(shape), np.nanstd(shape)])
        else:
            byamp_results.append([np.sum(group['numspots']),np.nanmean(flux), np.nanstd(flux), np.nanmean(shape), np.nanstd(shape)])

    
            
spots_pickle = {'results':byamp_results, 'corrected_results': byamp_corrected_results}
filename = "./plots/itl-R10_S02-correctedspots-moredata/polynomialfit/spots_results_12788_R10_S02.pkl"


if len(spots_pickle) == 0 or not (len(byamp_results) == len(byamp_corrected_results)):
    print("Error loading the data!")
else:
    if(os.path.isfile(filename)):
        print("Throwing out the old pickle.")
        os.remove(filename)

    with open(filename, 'wb') as f:
        print("Jarring a new pickle.")
        pkl.dump(spots_pickle, f)
        print("Done.")
```

```python
from astropy.table import vstack
col = "u/abrought/run_12781/bf_work_v5/polynomialfit/spots_v11/wbfcorr"

y1key="base_SdssShape_xx"
y2key="base_SdssShape_yy"
xkey="base_CircularApertureFlux_70_0_instFlux"
zskey="base_CircularApertureFlux_3_0_instFlux"
zskey="base_PsfFlux_instFlux"
zlkey="base_CircularApertureFlux_70_0_instFlux"

subbutler = Butler(repo_path,collections=col)
subregistry = subbutler.registry

datasetRefs = list(subregistry.queryDatasets(
    datasetType="gridSpotSrc",
    collections=col))

results = []

for aref in datasetRefs[:]:
    try:
        # get a catalog from the butler
        src = subbutler.get(aref)
        # get the associated metadata from raw data which contains EXPTIME
        raw = subbutler.get("raw",dataId=aref.dataId)
        img = raw.getImage().getArray()
        if img.min()<22000:
            raise
        md = raw.getMetadata()
        # stores into memory for later use
        results.append( {"md": md, "src":src} )
    except:
#        raise
        pass

faint = []
for aresult in results:
    faint.append(aresult["src"].asAstropy())
mean_data_table = vstack(faint, join_type='exact')

mean_data_table_by_groups = mean_data_table[ np.where(
        (mean_data_table["spotgrid_index"]>=0) &
        (mean_data_table["base_PixelFlags_flag"]==False) &
        (mean_data_table["base_PsfFlux_flag"]==False) &
        (mean_data_table["base_SdssCentroid_flag"]==False) &
        (mean_data_table["base_CircularApertureFlux_70_0_flag_apertureTruncated"]==False))].group_by('spotgrid_index')

ref = mean_data_table_by_groups.groups.aggregate(np.nanmedian)
```
